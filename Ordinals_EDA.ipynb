# Avg Ordian Rank

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the MMasseyOrdinals dataset
massey_ordinals = pd.read_csv('MMasseyOrdinals.csv')
```

```python
massey_ordinals.head()
```

```python
# Calculate metrics for each system
system_metrics = massey_ordinals.groupby(['Season', 'TeamID', 'SystemName']).agg(
    InitialRank=('OrdinalRank', 'first'),
    FinalRank=('OrdinalRank', 'last'),
    BestRank=('OrdinalRank', 'min'),
    WorstRank=('OrdinalRank', 'max'),
    MeanRank=('OrdinalRank', 'mean')
).reset_index()
```

```python
# Calculate Rank Improvement for each system
system_metrics['RankImprovement'] = system_metrics['InitialRank'] - system_metrics['FinalRank']
```

```python
# Aggregate metrics across systems
team_metrics = system_metrics.groupby(['Season', 'TeamID']).agg(
    AvgInitialRank=('InitialRank', 'mean'),
    AvgFinalRank=('FinalRank', 'mean'),
    AvgBestRank=('BestRank', 'mean'),
    AvgWorstRank=('WorstRank', 'mean'),
    AvgMeanRank=('MeanRank', 'mean'),
    AvgRankImprovement=('RankImprovement', 'mean')
).reset_index()
```

```python
# Display the resulting DataFrame
team_metrics.sample(20)
```
